# Generating our own data with random values:

```python
# Make empty data-frame to add random data to. 
car_data = pd.DataFrame()
```

```python
# Setup random seed
np.random.seed(seed=13)
random.seed(13)

# Make 1 line numpy array with 50 elements (these will be potential model years)
model_year_array = np.random.randint(1990, 2023, (50))

# Make array for car models 
potential_models = ["Toyota", "Nissan", "Honda", "Ferrari", "Audi", "BMW", "Subaru", "Jeep", "Tesla", "Ford", "Chrysler"]

models_array = np.random.choice(potential_models, 50)

# Make Odometer Data
odometer_array = np.random.randint(0, 50000, (50))


# Make pandas series' which will be used to construct dataframe 
year_series = pd.Series(model_year_array)
make_series = pd.Series(models_array)
odometer_series = pd.Series(odometer_array)

car_data["Year"] = year_series
car_data["Make"] = make_series 
car_data["Miles"] = odometer_series

# Lets make a price column and lets also make it dependent on the mileage 
car_data["Price"] = (50000 - (0.5 * car_data["Miles"])).astype(int)

# Lets make a Accident column to see if cars have been in an accident 
# Lets generate those bool values randomly but only have a 12% chance of a car having been in an accident 
probabilities = [0.12, 0.88]
accident_array = np.random.choice([True, False], size=50, p=probabilities)
accident_series = pd.Series(accident_array)
car_data["Collision History"] = accident_series

car_data
```

# Standard Deviation:

### Standard Deviation is used to find a value which represents how spread out data is 
### There are 2 methods for finding the standard deviation:
#### Method 1 applies when you have all of the data present 
#### Method 2 applies when the data represents a larger population 

## Method 1:




```python
sample_data = np.random.randint(0,1000, 300)
```

```python
sample_data
```

```python
def standard_dev(array):
    # find the average 
    mean = np.sum(sample_data) / np.size(sample_data)
    # find the sum of each numbers distance fromm the mean 
    distances = np.abs(sample_data - mean)
    # square each difference
    distances_squared = np.square(distances)
    # average of distances squared
    average_distances_squared = np.sum(distances_squared) / np.size(distances_squared)
    # standard devation = square root of the average of all distances squared
    standard_deviation = np.sqrt(average_distances_squared)
    return standard_deviation
```

```python
standard_dev(sample_data)
```

```python
# or more simply:
std = np.std(sample_data)
std
```

## Method 2 (Standard deviation on sample data)
### The only difference here is that when we get the average of the distances squared:
### instead of dividing by the number of values we divide by the number of values minus 1

```python
def standard_dev_sample(array):
    # find the average 
    mean = np.sum(sample_data) / np.size(sample_data)
    # find the sum of each numbers distance fromm the mean 
    distances = np.abs(sample_data - mean)
    # square each difference
    distances_squared = np.square(distances)
    # average of distances squared (get average with (actual size - 1) because this is standard dev of sample data
    average_distances_squared = np.sum(distances_squared) / (np.size(distances_squared) - 1)
    # standard devation = square root of the average of all distances squared
    standard_deviation = np.sqrt(average_distances_squared)
    return standard_deviation
```

```python
standard_dev_sample(sample_data)
```

```python
# or more simply:
np.std(sample_data, ddof=1)
```

# Dot Product
## finding the dot product of two vectors or matrices is useful to find similarities and patterns 

```python
plt.imshow(dot_product_pic)
plt.axis("off")
plt.show()
```

```python
# lets make two matrices and get the dot product 
m1 = np.random.randint(0,30, (5,3))
m2 = np.random.randint(0,30, (5,3))
```

```python
m1, m2
```

```python
np.dot(m1,m2)
```

#### The previous code throws an error
#### this is because in order to get the dot product of two matrices or vectors
#### the number of columns in the first matrix must be equal to the number of rows in the second matrix 
#### we can use the transpose of m2 to make the two shapes compatible 

```python
m2_transposed = np.transpose(m2)
output = f"Original Shapes: m1: {np.shape(m1)}, m2: {np.shape(m2)}, \nAfter Transpose: m1: {np.shape(m1)}, m2: {np.shape(m2_transposed)}"
print(output)
```

```python
# now that the shapes are compatible we can get the dot product of the two matrices 
np.dot(m1, m2_transposed)
```
