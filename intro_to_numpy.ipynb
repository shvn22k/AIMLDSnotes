```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
# Using Matplotlib to display saved images
numpy1 = mpimg.imread("numpy1.png")
numpy2 = mpimg.imread("numpy2.png")
numpy3 = mpimg.imread("numpy3.png")
plt.imshow(numpy1)
plt.axis("off")
plt.show()
plt.imshow(numpy2)
plt.axis("off")
plt.show()
plt.imshow(numpy3)
plt.axis("off")
plt.show()
```

### Data Types & Attributes:

```python
# The main data-type in numpy is called ndarray which stands for (n dimensional array) 
# Here is a 1 dimensional array
a1 = np.array([1,2,3])
type(a1)
```

```python
a1
```

```python
# Here is a 2 dimensional array 
a2 = np.array([[1,2,3],[4, 3.4, 9]])
a2
```

```python
# Here is a 3 dimensional array 
a3 = np.array([[[1, 2, 3, 32],[4, 5, 6, 32],[7, 8, 9, 32]],[[10,11,12,32],[13,14,15,32],[16,17,18,32]]])
a3
```

#### ndarray's can be any size, to view the shape of an array use the shape method

```python
# 3 columns
a1.shape
```

```python
# 2 rows 3 columns 
a2.shape
```

```python
# 2 arrays with 3 rows and 4 columns
a3.shape
```

```python
# use ndim to show number of dimensions in an array
a1.ndim, a2.ndim, a3.ndim
```
